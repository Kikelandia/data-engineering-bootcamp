```python
sampleLocation = "Datos/sample.csv"
sampleData = pd.read_csv(sampleLocation)
```

```python
fullDataLocation = "Datos/profeco/all_data.csv"
chunkData = pd.read_csv(fullDataLocation,
                        usecols=['cadenaComercial','estado','producto'],
                        chunksize=1000000,
                        iterator=True,
                        header=0,
                        low_memory=True)
fullData = pd.concat(chunkData)
```

## How many commercial chains are monitored, and therefore, included in this database?

```python
print(sampleData.cadenaComercial.unique())
print(sampleData.cadenaComercial.nunique())
```

```python
def commChainsMonitored(dataFrame):
    result = []
    for item in dataFrame.cadenaComercial:
        if item in result:
            continue
        else:
            result.append(item)
    return result
```

```python
cadenas_Monitoreadas_Sample = commChainsMonitored(sampleData)
print(cadenas_Monitoreadas_Sample)
print(len(cadenas_Monitoreadas_Sample))
```

```python
print(fullData.cadenaComercial.unique())
print(fullData.cadenaComercial.nunique())
```

```python
cadenas_Monitoreadas_Full = commChainsMonitored(fullData)
print(cadenas_Monitoreadas_Full)
print(len(cadenas_Monitoreadas_Full))
```

## What are the lessons learned from this exercise?

A good start option is working with a slice of the data, I believe a randomized slice would fit best a first approach so we can have a more diverse data

Luckily we had no nan values in the full data I have used however, this almost never happens we will need to learn data tricks to handle inconsistent data

If is necessary to work locally with data, one good option is to load exclusively the data we need to work with, storing all data in memory will decrease the speed and performance of our kernel/machine/procesor
